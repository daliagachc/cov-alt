```python
f = './data/south.america.subnational.mortality.txt'

df = pd.read_csv(f)
```

```python
df['dt']=pd.to_datetime(df['date'])
df = df.set_index('dt')
df = df['2019':'2022']
```

```python
n = df['adm1_isocode'].drop_duplicates()
```

```python
for i in n[:]:
    f,ax = plt.subplots(figsize=(3,1))
    bo = df['adm1_isocode'] == i 
    _d = df[bo]
    dd = _d['deaths']
    dd.resample('M').mean().plot(zorder=100)
    _d['deaths'].plot()
    ax.set_xlim('2019','2021-07')
    ax.set_title(i)
```
